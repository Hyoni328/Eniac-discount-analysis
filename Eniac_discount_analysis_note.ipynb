#### **How do seasonality and special dates (Christmas, Black Friday) affect sales?**

```python
orders_df.loc[:, 'month'] = orders_df['created_date'].dt.month
orders_df.head()
```

```python
orders_df.loc[:, 'year'] = orders_df['created_date'].dt.year
```

```python
orders_df.loc[:, 'month_year'] = orders_df['created_date'].dt.strftime('%b, %Y')
orders_df.head()
```

##### Monthly

```python
monthly_sales = orders_df.groupby('month_year')['total_paid'].sum()
monthly_sales = monthly_sales.reset_index()
monthly_sales['month_year'] = pd.to_datetime(monthly_sales['month_year'])
monthly_sales = monthly_sales.sort_values('month_year')
```

```python
monthly_sales = monthly_sales.reset_index()
```

```python
monthly_sales = monthly_sales.drop('index', axis=1)
```

```python
sales_2017 = monthly_sales.iloc[:12]
sales_2017
```

```python
monthly_sales.loc[(0, 12), :]
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
sns.set(font_scale=2)
sns.set_style('white')

(
sns.relplot(kind='line',
            data=monthly_sales,
            x='month_year',
            y='total_paid',
            height=8,
            aspect=2)
.set_axis_labels('', 'Total Sales (Million €)')
);
```

##### Season

```python
# Season
orders_df['season'] = ''
orders_df.loc[(orders_df['month'].isin([1,2])) & (orders_df['year'] == 2017), 'season'] = '17 Winter'
orders_df.loc[(orders_df['month'].isin([3,4,5])) & (orders_df['year'] == 2017), 'season'] = '17 Spring'
orders_df.loc[orders_df['month'].isin([6,7,8]), 'season'] = '17 Summer'
orders_df.loc[orders_df['month'].isin([9,10,11]), 'season'] = '17 Autumn'
orders_df.loc[orders_df['month'].isin([12,1,2]) & (orders_df['season'] ==''), 'season'] = '18 Winter'
orders_df.loc[(orders_df['month'].isin([3])) & (orders_df['year'] == 2018), 'season'] = '18 Spring'
```

```python
orders_df['season_order'] = ''
orders_df.loc[(orders_df['season'] == '17 Winter'), 'season_order'] = 1
orders_df.loc[(orders_df['season'] == '17 Spring'), 'season_order'] = 2
orders_df.loc[(orders_df['season'] == '17 Summer'), 'season_order'] = 3
orders_df.loc[(orders_df['season'] == '17 Autumn'), 'season_order'] = 4
orders_df.loc[(orders_df['season'] == '18 Winter'), 'season_order'] = 5
orders_df.loc[(orders_df['season'] == '18 Spring'), 'season_order'] = 6
```

```python
season_sales = orders_df.groupby(['season', 'season_order'])['total_paid'].sum()
season_sales = season_sales.reset_index()
season_sales = season_sales.sort_values('season_order')
```

```python
sns.set(font_scale=1)

(
sns.relplot(kind='line',
            data=season_sales,
            x='season',
            y='total_paid',
            height=4,
            aspect=2)
);
```

##### Christmas

```python
#Christmas
dec_sales = orders_df.loc[orders_df['month'] == 12]
dec_sales.insert(4, 'day', '')
dec_sales.loc[:, 'day'] = dec_sales['created_date'].dt.day
dec_sales.head()
```

```python
dec_daily_sales = dec_sales.groupby('day')['total_paid'].sum()
dec_daily_sales = dec_daily_sales.reset_index()
```

#### **Top 10 products**

```python
top_10_sold = ol_merge.groupby('name')[['product_quantity', 'unit_price_total']].sum().sort_values('product_quantity', ascending= False).head(10)
```

```python
top_10_sales = ol_merge.groupby('name')[['product_quantity', 'unit_price_total']].sum().sort_values('unit_price_total', ascending= False).head(10)
```

```python
sns.set(font_scale=1)

(
sns.catplot(kind="bar",
            data=top_10_sold,
            x="product_quantity",
            y="name",
            color="#78c679",
            aspect=3,
            sharey=False)
    .figure.suptitle("Top ten products by sold units")
    .figure.subplots_adjust(top=.92)
);
```

```python
sns.set(font_scale=1)

(
sns.catplot(kind="bar",
            data=top_10_sales,
            x="unit_price_total",
            y="name",
            color="#addd8e",
            aspect=3,
            sharey=False)
    .figure.suptitle("Top ten products by sales")
    .figure.subplots_adjust(top=.92)
);
```
